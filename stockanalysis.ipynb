```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

msft=yf.Ticker('msft')
```

```python
stockinfo=msft.info
```

```python
for key,value in stockinfo.items():
    print(key,':',value)
```

```python
df=msft.dividends
```

```python
data=df.resample('Y').sum()
data=data.reset_index()
```

```python
data['Year']=data['Date'].dt.year
plt.figure()
plt.bar(data['Year'],data['Dividends'])
plt.ylabel('Dividend Yield($)')
plt.xlabel('Year')
plt.title('Microsoft dividend history')
plt.xlim(2002,2023)
plt.show()
```

```python
#from the above plot it is interesting to note that microsoft had issued a dividend in 2004 which was much higher than its historical average
#and also it has been consistently increasing its dividend every year
```

```python
#Visual representation of the performance of different securities from the year 2000 to 2024
```

```python
securities=['on','msft','aapl','tsla','nvda','amzn','meta','goog']
df2=pd.DataFrame()
today=datetime.now().date().strftime("%Y-%m-%d")
```

```python
for security in securities:
    df2[security]=yf.Ticker(security).history(start='2000-01-01',end=today).Close
```

```python
plt.figure()
lines=plt.plot(df2,label="securities")
plt.legend(lines,securities)
plt.ylabel('Price($)')
plt.xlabel('Year')
plt.show()
```
